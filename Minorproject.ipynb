```python
import pandas as pd

df = pd.read_csv("/content/spotify dataset.csv.csv")
df
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Convert relevant columns to numeric types
df['duration_ms'] = pd.to_numeric(df['duration_ms'], errors='coerce')
df['track_popularity'] = pd.to_numeric(df['track_popularity'], errors='coerce')
```

```python
# Drop rows with missing values
df_cleaned = df.dropna()
```

```python
# Descriptive statistics
df_cleaned.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define numerical columns
numerical_cols = ['danceability', 'energy', 'key', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']

# Compute the correlation matrix
corr_matrix = df_cleaned[numerical_cols].corr()

# Plot the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Visualize the distribution of energy, danceability, and loudness
plt.figure(figsize=(10, 6))
sns.histplot(data=df_cleaned, x='danceability', kde=True, color='blue', label='Danceability')
sns.histplot(data=df_cleaned, x='energy', kde=True, color='red', label='Energy')
sns.histplot(data=df_cleaned, x='loudness', kde=True, color='green', label='Loudness')
plt.legend()
plt.title('Distribution of Danceability, Energy, and Loudness')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# Standardizing the numerical features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_cleaned[numerical_cols])

# Convert the scaled data back into a DataFrame
scaled_data_df = pd.DataFrame(scaled_data, columns=numerical_cols)
```

```python
from sklearn.cluster import KMeans

# Apply KMeans clustering
kmeans = KMeans(n_clusters=5, random_state=42)

# Create a copy of the DataFrame before assigning the cluster column
df_cleaned_copy = df_cleaned.copy()
df_cleaned_copy.loc[:, 'cluster'] = kmeans.fit_predict(scaled_data_df)

# Update df_cleaned to refer to the copied DataFrame with the cluster column
df_cleaned = df_cleaned_copy

# Check the first few rows with cluster assignments
df_cleaned[['track_name', 'cluster']].head()
```

```python
# Visualize the clusters using energy vs danceability
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_cleaned, x='energy', y='danceability', hue='cluster', palette='Set1', s=100)
plt.title('Clusters of Songs Based on Energy and Danceability')
plt.show()
```

```python
def recommend_songs(liked_song_id):
    # Find the cluster of the liked song
    liked_song_cluster = df_cleaned[df_cleaned['track_id'] == liked_song_id]['cluster'].values[0]

    # Get all songs in the same cluster
    recommended_songs = df_cleaned[df_cleaned['cluster'] == liked_song_cluster]

    # Return recommended songs with track name, artist, and cluster info
    return recommended_songs[['track_name', 'track_artist', 'cluster']].head(10)

# Example: Recommend songs for a given song (using track_id from the dataset)
liked_song_id = '6f807x0ima9a1j3VPbc7VN'  # Replace with any song ID from the dataset
recommended_songs = recommend_songs(liked_song_id)

print(recommended_songs)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Select the numerical features for similarity calculation
features = ['danceability', 'energy', 'loudness', 'tempo', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence']

# Extract the feature data
feature_data = df_cleaned[features]
```

```python
from sklearn.preprocessing import StandardScaler

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(feature_data)
```

```python
# Compute the cosine similarity matrix
cosine_sim = cosine_similarity(scaled_features)

# Create a DataFrame to store the similarity values
cosine_sim_df = pd.DataFrame(cosine_sim, index=df_cleaned['track_name'], columns=df_cleaned['track_name'])
```

```python
def recommend_similar_songs(song_name, num_recommendations=10):
    # Check if the song exists in the dataset
    if song_name not in cosine_sim_df.index:
        return "Song not found in the dataset."

    # Get the similarity scores for the given song, handling potential duplicates
    # Select the row(s) where the index matches song_name and take the first one
    sim_scores = cosine_sim_df.loc[cosine_sim_df.index == song_name].iloc[0]


    # Sort the songs based on similarity score (descending order)
    sim_scores_sorted = sim_scores.sort_values(ascending=False)

    # Get the top N recommendations (excluding the song itself)
    recommended_songs = sim_scores_sorted[1:num_recommendations+1]

    # Return the recommended song names and their similarity scores
    return recommended_songs

# Example: Recommend songs similar to 'I Don't Care (with Justin Bieber) - Loud Luxury Remix'
recommended_songs = recommend_similar_songs('I Don\'t Care (with Justin Bieber) - Loud Luxury Remix', 10)

print(recommended_songs)
```
